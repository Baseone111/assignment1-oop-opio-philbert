```python
#for loop to interate through the train_dataset to identify the percentage of null values in some columns
for i in train_dataset.columns:
    null_rate = train_dataset[i].isna().sum() / len(train_dataset) * 100
    if null_rate > 0 :
        print("{} null rate: {}%".format(i,round(null_rate,2)))
```

```python
#fillining in null values for the three columns that really have a big percentage of nulls and drop some rows of the other columns that have

train_dataset['country'] = train_dataset['country'].fillna(train_dataset['country'].mode()[0])


train_dataset['cast'].replace(np.nan, 'No Data',inplace  = True)
train_dataset['director'].replace(np.nan, 'No Data',inplace  = True)

# Drops

train_dataset.dropna(inplace=True)

# Drop Duplicates

train_dataset.drop_duplicates(inplace= True)
```

```python
# Define the Show class
class Show:
    def __init__(self, show_id, show_type, title, director, cast, country, date_added, release_year, rating, duration, listed_in, description):
        self.show_id = show_id
        self.type = show_type
        self.title = title
        self.director = director
        self.cast = cast
        self.country = country
        self.date_added = date_added
        self.release_year = release_year
        self.rating = rating
        self.duration = duration
        self.listed_in = listed_in
        self.description = description

    # Method to determine if it's family-friendly
    def is_family_friendly(self):
        family_ratings = ["G", "PG", "PG-13", "TV-Y", "TV-G", "TV-Y7", "TV-PG"]
        return self.rating in family_ratings
```

```python
# Function to filter only family-friendly content
def filter_family_friendly(shows):
        friendly_list = []
        for show in shows:
            if show.is_family_friendly():
                friendly_list.append(show)
        return friendly_list
```

```python

# Function to count shows by genre (listed_in column)
def count_by_genre(shows):
    genre_count = {}
    for show in shows:
        genres = show.listed_in.split(", ")  # Some shows have multiple genres
        for genre in genres:
            if genre in genre_count:
                genre_count[genre] += 1
            else:
                genre_count[genre] = 1
    return genre_count
```

```python
# Create a list of Show objects from the DataFrame
shows_list = []
for index, row in train_dataset.iterrows():
    show = Show(
        row['show_id'], row['type'], row['title'], row['director'],
        row['cast'], row['country'], row['date_added'], row['release_year'],
        row['rating'], row['duration'], row['listed_in'], row['description']
    )
    shows_list.append(show)

family_friendly = filter_family_friendly(shows_list)
genge_stats = count_by_genre(shows_list)
```

```python
print("ðŸŽ¬ FAMILY-FRIENDLY SHOWS:")
for show in family_friendly:
    print(f"- {show.title} ({show.rating})")

print("\nðŸ“Š GENRE COUNTS:")
for genre, count in genre_stats.items():
    print(f"{genre}: {count} shows")
```

```python
# a) Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# b) Generate prime numbers between 1 and 50
prime_numbers = [n for n in range(1, 51) if is_prime(n)]

# c) Display prime numbers on a single line
print("Prime numbers between 1 and 50:", *prime_numbers)
```

```python
import math

# a) Define Coordinate class
class Coordinate:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # b) Method to compute Euclidean distance
    def distance(self, other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

# c) Prompt user for coordinates
# Get coordinates for first point
x1 = float(input("Enter x coordinate for point 1: "))
y1 = float(input("Enter y coordinate for point 1: "))
point1 = Coordinate(x1, y1)

# Get coordinates for second point
x2 = float(input("Enter x coordinate for point 2: "))
y2 = float(input("Enter y coordinate for point 2: "))
point2 = Coordinate(x2, y2)

# Print the distance
print(f"Distance between ({point1.x},{point1.y}) and ({point2.x},{point2.y}): {point1.distance(point2)}")
```
